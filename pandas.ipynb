```python
# pivot
# Next, we create a sample DataFrame sales_data with sales information.
data={
    'Date': ['2024-05-01', '2024-05-01', '2024-05-02', '2024-05-02'],
    'Product': ['A', 'B', 'A', 'B'],
    'Revenue': [100, 200, 150, 250]
}
```

```python
# We display the original DataFrame to see the structure of the data.
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
# After pivoting, each unique date becomes a row index, each unique product becomes a column, 
# and the corresponding revenue values are filled in the table.

df.pivot(index="Date",columns="Product",values="Revenue")
```

```python
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Values': [10, 20, 15, 25]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Creating a pivot table
pd.pivot_table(df, values='Values', index='Date', columns='Category', aggfunc='sum')
```
