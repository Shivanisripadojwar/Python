```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the figure
fig = plt.figure()

# Define the GridSpec
gs = gridspec.GridSpec(3, 3)

# Add subplots
ax1 = fig.add_subplot(gs[0, :])      # First row, spans all columns
ax2 = fig.add_subplot(gs[1, :-1])    # Second row, first two columns
ax3 = fig.add_subplot(gs[1:, -1])    # Second and third row, last column
ax4 = fig.add_subplot(gs[-1, 0])     # Last row, first column
ax5 = fig.add_subplot(gs[-1, -2])    # Last row, middle column

# Plot data on each subplot
ax1.plot(x, y1, label='sin(x)')
ax1.set_title('Plot 1')
ax1.legend()

ax2.plot(x, y2, label='cos(x)', color='orange')
ax2.set_title('Plot 2')
ax2.legend()

ax3.plot(x, -y1, label='-sin(x)', color='green')
ax3.set_title('Plot 3')
ax3.legend()

ax4.plot(x, -y2, label='-cos(x)', color='red')
ax4.set_title('Plot 4')
ax4.legend()

ax5.plot(x, y1 - y2, label='sin(x) - cos(x)', color='purple')
ax5.set_title('Plot 5')

```

```python
# Advanced Techniques
# 1. Animations

import matplotlib.animation as animation

fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))

def update(num, x, line):
    line.set_ydata(np.sin(x + num / 10.0))
    return line,

ani = animation.FuncAnimation(fig, update, frames=100, fargs=[x, line], blit=True)
plt.show()
```

```python
# Stackplot- hat accumulates over time, showing how different parts contribute to the whole
import matplotlib.pyplot as plt

# Data
years = [2018, 2019, 2020, 2021, 2022]
category1 = [2, 3, 5, 7, 6]
category2 = [1, 2, 4, 5, 4]
category3 = [2, 3, 6, 8, 7]

# Create the stack plot with customization
plt.stackplot(years, category1, category2, category3, colors=['#ff9999','#66b3ff','#99ff99'], alpha=0.8, labels=['Category 1', 'Category 2', 'Category 3'])

# Add title and labels
plt.title('Customized Stack Plot')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
# violin plot lotting numeric data and can be considered a combination of a box plot and a kernel density plot.
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data
np.random.seed(10)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 2, 100)
data3 = np.random.normal(2, 1.5, 100)
data = [data1, data2, data3]

# Create the violin plot with customization
sns.violinplot(data=data, palette='Set3', inner='quartile', orient='v')

# Add title and labels
plt.title('Customized Violin Plot')
plt.xlabel('Category')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
# 3d lineplot
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
t = np.linspace(0, 20, 1000)
x = np.sin(t)
y = np.cos(t)
z = t

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)

# Add title and labels
ax.set_title('3D Line Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show the plot
plt.show()
```
